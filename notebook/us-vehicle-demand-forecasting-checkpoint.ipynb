```python
import pandas as pd
import numpy as np

# =========================
# CONFIG
# =========================
FILE_PATH = r"C:\Users\omogun01\Story telling with Data\BEA Auto-Light Truck Sales.xlsx"
OUT_PATH  = r"C:\Users\omogun01\Story telling with Data\vehicle_sales_sa_monthly.csv"

AUTO_SHEET  = "Auto Sales (1990-2016)"
TRUCK_SHEET = "Light Truck Sales (1990-2016)"

# =========================
# HELPERS
# =========================
MONTH_MAP = {
    "January": 1, "February": 2, "March": 3, "April": 4,
    "May": 5, "June": 6, "July": 7, "August": 8,
    "September": 9, "October": 10, "November": 11, "December": 12
}

def clean_bea_sheet(df: pd.DataFrame, prefix: str) -> pd.DataFrame:
    """
    Extract seasonally adjusted (Thousands) domestic + foreign series from BEA sheet.
    Returns a tidy dataframe with: date, <prefix>_domestic_sa, <prefix>_foreign_sa
    """
    # Keep only the actual data rows (Month not null)
    df = df.copy()
    df = df[df["Unnamed: 0"].notna()]

    # Rename the needed columns
    out = df.rename(columns={
        "Unnamed: 0": "month_name",
        "Unnamed: 1": "year",
        "Unnamed: 4": f"{prefix}_domestic_sa_thousands",
        "Unnamed: 9": f"{prefix}_foreign_sa_thousands",
    })[
        ["month_name", "year", f"{prefix}_domestic_sa_thousands", f"{prefix}_foreign_sa_thousands"]
    ]

    # Standardize types
    out["month_name"] = out["month_name"].astype(str).str.strip()
    out["year"] = pd.to_numeric(out["year"], errors="coerce")

    # Convert month name -> month number
    out["month"] = out["month_name"].map(MONTH_MAP)

    # Drop any non-month rows that slipped in
    out = out.dropna(subset=["year", "month"])

    # Create a proper monthly date
    out["date"] = pd.to_datetime(
        dict(year=out["year"].astype(int), month=out["month"].astype(int), day=1)
    )

    # Final cleanup
    out = out.drop(columns=["month_name", "month", "year"]).sort_values("date").reset_index(drop=True)

    # Ensure numeric
    for c in [f"{prefix}_domestic_sa_thousands", f"{prefix}_foreign_sa_thousands"]:
        out[c] = pd.to_numeric(out[c], errors="coerce")

    return out

# =========================
# LOAD
# =========================
auto_raw  = pd.read_excel(FILE_PATH, sheet_name=AUTO_SHEET)
truck_raw = pd.read_excel(FILE_PATH, sheet_name=TRUCK_SHEET)

# =========================
# CLEAN (Seasonally Adjusted only)
# =========================
auto_sa  = clean_bea_sheet(auto_raw,  prefix="auto")
truck_sa = clean_bea_sheet(truck_raw, prefix="truck")

# =========================
# MERGE
# =========================
df = auto_sa.merge(truck_sa, on="date", how="inner")

# Derived totals (still in thousands)
df["auto_total_sa_thousands"]   = df["auto_domestic_sa_thousands"] + df["auto_foreign_sa_thousands"]
df["truck_total_sa_thousands"]  = df["truck_domestic_sa_thousands"] + df["truck_foreign_sa_thousands"]
df["vehicle_total_sa_thousands"]= df["auto_total_sa_thousands"] + df["truck_total_sa_thousands"]

# Market shares (avoid divide by zero)
df["auto_share"]  = np.where(df["vehicle_total_sa_thousands"] > 0,
                             df["auto_total_sa_thousands"] / df["vehicle_total_sa_thousands"], np.nan)
df["truck_share"] = np.where(df["vehicle_total_sa_thousands"] > 0,
                             df["truck_total_sa_thousands"] / df["vehicle_total_sa_thousands"], np.nan)

# =========================
# QUICK QUALITY CHECKS
# =========================
print("Rows:", len(df))
print("Date range:", df["date"].min().date(), "→", df["date"].max().date())
print("\nMissing values per column:\n", df.isna().sum())

# =========================
# EXPORT
# =========================
df.to_csv(OUT_PATH, index=False)
print("\nSaved:", OUT_PATH)

df.head()
```

## Peak-to-Trough Decline Calculation

```python
def peak_to_trough(series):
    peak = series.max()
    trough = series.min()
    decline_pct = (trough - peak) / peak
    return peak, trough, decline_pct

auto_peak, auto_trough, auto_decline = peak_to_trough(recession_window["auto_total_sa_thousands"])
truck_peak, truck_trough, truck_decline = peak_to_trough(recession_window["truck_total_sa_thousands"])

print("Auto decline:", round(auto_decline * 100, 2), "%")
print("Truck decline:", round(truck_decline * 100, 2), "%")
```

## Recovery Duration

```python
def recovery_time(series, peak_value):
    after_trough = series[series.idxmin():]
    recovered = after_trough[after_trough >= peak_value]
    if len(recovered) > 0:
        return (recovered.index[0] - series.idxmin())
    else:
        return None

auto_recovery = recovery_time(recession_window["auto_total_sa_thousands"], auto_peak)
truck_recovery = recovery_time(recession_window["truck_total_sa_thousands"], truck_peak)

print("Auto recovery months:", auto_recovery)
print("Truck recovery months:", truck_recovery)
```

## Stationarity Test (ADF)--- This checkes whether differencing is needed 

```python
def adf_test(series, name=""):
    result = adfuller(series.dropna())
    print(f"ADF Test for {name}")
    print("ADF Statistic:", result[0])
    print("p-value:", result[1])
    print("-" * 40)

targets = [
    "auto_domestic_sa_thousands",
    "auto_foreign_sa_thousands",
    "truck_domestic_sa_thousands",
    "truck_foreign_sa_thousands"
]

for t in targets:
    adf_test(train[t], name=t)
```

## Generic SARIMA Function

```python
def fit_sarima(train_series, test_series, 
               order=(1,1,1), 
               seasonal_order=(1,1,1,12)):
    
    model = SARIMAX(
        train_series,
        order=order,
        seasonal_order=seasonal_order,
        enforce_stationarity=False,
        enforce_invertibility=False
    )
    
    results = model.fit(disp=False)
    
    forecast = results.forecast(steps=len(test_series))
    forecast = forecast.reindex(test_series.index)
    
    # Safe evaluation
    valid_idx = (~forecast.isna()) & (~test_series.isna())
    
    mae = mean_absolute_error(test_series[valid_idx], forecast[valid_idx])
    rmse = np.sqrt(mean_squared_error(test_series[valid_idx], forecast[valid_idx]))
    mape = np.mean(np.abs((test_series[valid_idx] - forecast[valid_idx]) / test_series[valid_idx])) * 100
    
    return {
        "model": results,
        "forecast": forecast,
        "MAE": mae,
        "RMSE": rmse,
        "MAPE": mape
    }
```

```python
results_dict = {}

for target in targets:
    print(f"Fitting SARIMA for {target}")
    res = fit_sarima(train[target], test[target])
    results_dict[target] = res
```

```python
def train_ml_model(model, X_train, y_train, X_test, y_test):
    
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    
    mae = mean_absolute_error(y_test, preds)
    rmse = np.sqrt(mean_squared_error(y_test, preds))
    mape = np.mean(np.abs((y_test - preds) / y_test)) * 100
    
    return {
        "model": model,
        "predictions": preds,
        "MAE": mae,
        "RMSE": rmse,
        "MAPE": mape
    }
```

## TimeSeries Cross-Validation

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error
import numpy as np

def cross_validate_model(model, X, y, n_splits=5):
    
    tscv = TimeSeriesSplit(n_splits=n_splits)
    rmses = []
    
    for train_idx, val_idx in tscv.split(X):
        
        # Skip very small training windows
        if len(train_idx) < 60:   # require at least 5 years history
            continue
        
        X_train_cv, X_val_cv = X.iloc[train_idx], X.iloc[val_idx]
        y_train_cv, y_val_cv = y.iloc[train_idx], y.iloc[val_idx]
        
        model.fit(X_train_cv, y_train_cv)
        preds = model.predict(X_val_cv)
        
        rmse = np.sqrt(mean_squared_error(y_val_cv, preds))
        rmses.append(rmse)
        
    return np.mean(rmses)
```

```python
cv_results = []

for target in targets:
    X = df[feature_cols]
    y = df[target]
    
    model = LinearRegression()
    cv_rmse = cross_validate_model(model, X, y)
    
    cv_results.append({
        "Segment": target,
        "CV_RMSE": cv_rmse
    })

cv_df = pd.DataFrame(cv_results)
cv_df
```
